```python
import requests
from bs4 import BeautifulSoup

page = requests.get('https://www.yallakora.com/match-center/مركز-المباريات?date=8/4/2024#')
matches_details=[]
from google.colab import drive
drive.mount('/content/drive')

def main():

    src = page.content  # string of html
    soup = BeautifulSoup(src, "lxml")  # parse from string to html
    #print(soup)   # html
    champions = soup.find_all("div", {'class': 'matchesList'})  # list of html code
    for champion in champions:


        title = champion.contents[1].find("h2").text.strip()   # strip remove the spaces in the beginign and in the end
        print('===================== Champion : ', title, '===================================')

        liste_matches = champion.contents[3].find_all("div", {'class': 'item'})
        print(len(liste_matches))
        i=1
        for matche in liste_matches:

            print('=== Matche : ', i, '=====')

            alldata = matche.find('a').find('div', {'class': 'allData'})
            date = alldata.find('div', {'class': 'topData'}).find('div', {'class': 'date'}).text.strip()
            teamA = alldata.find('div', {'class': 'teamCntnr'}).find('div', {'class': 'teamsData'}).find('div', {'class': 'teamA'}).find('p').text.strip()
            teamB = alldata.find('div', {'class': 'teamCntnr'}).find('div', {'class': 'teamsData'}).find('div', {'class': 'teamB'}).find('p').text.strip()
            time = alldata.find('div', {'class': 'teamCntnr'}).find('div', {'class': 'teamsData'}).find('div', {'class': 'MResult'}).find('span', {'class': 'time'}).text.strip()
            score_teamA = alldata.find('div', {'class': 'teamCntnr'}).find('div', {'class': 'teamsData'}).find('div', {'class': 'MResult'}).contents[1].text.strip()
            if(score_teamA =='-') :
              score_teamA='0'

            score_teamB = alldata.find('div', {'class': 'teamCntnr'}).find('div', {'class': 'teamsData'}).find('div', {'class': 'MResult'}).contents[3].text.strip()
            if(score_teamB =='-') :
              score_teamB='0'

            print("date= ", date)
            print("team A " ,teamA)
            print("team B= ", teamB)
            print("time= ", time)
            print("score team ", teamA, 'is : ', score_teamA)
            print("score team ", teamB, 'is : ', score_teamB)
            i=i+1
            matches_details.append({
                'champion_type: ': title,
                'team A' : teamA,
                'team B' : teamB,
                'time' : time,
                'score' : f'{score_teamA} - {score_teamB}'

            })

    keys= matches_details[0].keys()
    with open('/content/drive/My Drive/matches-details.csv', 'w') as output_file :
      dict_writter= csv.DictWriter(output_file, keys)
      dict_writter.writeheader()
      dict_writter.writerows(matches_details)
      print('file created')
main()
```
